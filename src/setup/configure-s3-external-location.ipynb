# Configure storage credential

1. Create a storage credential 
2. Test the connection

```python
%sql
DROP EXTERNAL LOCATION IF EXISTS `tpcds_lakehouse`;

CREATE EXTERNAL LOCATION IF NOT EXISTS tpcds_lakehouse
URL 's3://satyaki-databricks-lakehouse/tpcds'
WITH (STORAGE CREDENTIAL `satyaki-s3-sc`)
COMMENT 'creating tpcds lakehouse external location';
```

```python
%sql

CREATE CATALOG IF NOT EXISTS tpcds_lakehouse
      MANAGED LOCATION 's3://satyaki-databricks-lakehouse/tpcds'
      COMMENT 'This is the catalog for the TPC-DS Data Lakehouse' ;
```

```python
%sql

USE CATALOG tpcds_lakehouse;


CREATE SCHEMA IF NOT EXISTS landing
     MANAGED LOCATION 's3://satyaki-databricks-lakehouse/tpcds/landing';  
CREATE SCHEMA IF NOT EXISTS bronze
     MANAGED LOCATION 's3://satyaki-databricks-lakehouse/tpcds/bronze';  
CREATE SCHEMA IF NOT EXISTS silver
     MANAGED LOCATION 's3://satyaki-databricks-lakehouse/tpcds/silver';  
CREATE SCHEMA IF NOT EXISTS gold
     MANAGED LOCATION 's3://satyaki-databricks-lakehouse/tpcds/gold';   
```

```python
%sql
CREATE EXTERNAL VOLUME IF NOT EXISTS tpcds_lakehouse.landing.raw_data
LOCATION 's3://satyaki-databricks-lakehouse/tpcds/landing/data'
COMMENT 'TPC-DS landing volume'
```

```python
%sql
SELECT *
FROM read_files(
  's3://satyaki-databricks-lakehouse/tpcds/landing/data/customer.dat',
  format => 'csv',
  delimiter => '|',
  header => 'false'
);
```
