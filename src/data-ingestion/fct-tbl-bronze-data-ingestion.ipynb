```sql
-- ── CREATE TABLE ──────────────────────────────────────────────
CREATE TABLE IF NOT EXISTS bronze.store_sales (
    ss_sold_date_sk           INTEGER,
    ss_sold_time_sk           INTEGER,
    ss_item_sk                INTEGER,
    ss_customer_sk            INTEGER,
    ss_cdemo_sk               INTEGER,
    ss_hdemo_sk               INTEGER,
    ss_addr_sk                INTEGER,
    ss_store_sk               INTEGER,
    ss_promo_sk               INTEGER,
    ss_ticket_number          BIGINT,
    ss_quantity               INTEGER,
    ss_wholesale_cost         DECIMAL(7,2),
    ss_list_price             DECIMAL(7,2),
    ss_sales_price            DECIMAL(7,2),
    ss_ext_discount_amt       DECIMAL(7,2),
    ss_ext_sales_price        DECIMAL(7,2),
    ss_ext_wholesale_cost     DECIMAL(7,2),
    ss_ext_list_price         DECIMAL(7,2),
    ss_ext_tax                DECIMAL(7,2),
    ss_coupon_amt             DECIMAL(7,2),
    ss_net_paid               DECIMAL(7,2),
    ss_net_paid_inc_tax       DECIMAL(7,2),
    ss_net_profit             DECIMAL(7,2)
)
USING DELTA;

-- ── COPY INTO ─────────────────────────────────────────────────
COPY INTO bronze.store_sales
FROM (
    SELECT
        CAST(_c0  AS INTEGER)      AS ss_sold_date_sk,
        CAST(_c1  AS INTEGER)      AS ss_sold_time_sk,
        CAST(_c2  AS INTEGER)      AS ss_item_sk,
        CAST(_c3  AS INTEGER)      AS ss_customer_sk,
        CAST(_c4  AS INTEGER)      AS ss_cdemo_sk,
        CAST(_c5  AS INTEGER)      AS ss_hdemo_sk,
        CAST(_c6  AS INTEGER)      AS ss_addr_sk,
        CAST(_c7  AS INTEGER)      AS ss_store_sk,
        CAST(_c8  AS INTEGER)      AS ss_promo_sk,
        CAST(_c9  AS BIGINT)       AS ss_ticket_number,
        CAST(_c10 AS INTEGER)      AS ss_quantity,
        CAST(_c11 AS DECIMAL(7,2)) AS ss_wholesale_cost,
        CAST(_c12 AS DECIMAL(7,2)) AS ss_list_price,
        CAST(_c13 AS DECIMAL(7,2)) AS ss_sales_price,
        CAST(_c14 AS DECIMAL(7,2)) AS ss_ext_discount_amt,
        CAST(_c15 AS DECIMAL(7,2)) AS ss_ext_sales_price,
        CAST(_c16 AS DECIMAL(7,2)) AS ss_ext_wholesale_cost,
        CAST(_c17 AS DECIMAL(7,2)) AS ss_ext_list_price,
        CAST(_c18 AS DECIMAL(7,2)) AS ss_ext_tax,
        CAST(_c19 AS DECIMAL(7,2)) AS ss_coupon_amt,
        CAST(_c20 AS DECIMAL(7,2)) AS ss_net_paid,
        CAST(_c21 AS DECIMAL(7,2)) AS ss_net_paid_inc_tax,
        CAST(_c22 AS DECIMAL(7,2)) AS ss_net_profit
    FROM 's3://satyaki-databricks-lakehouse/tpcds/landing/data/store_sales.dat'
)
FILEFORMAT = CSV
FORMAT_OPTIONS ('delimiter' = '|', 'header' = 'false')
COPY_OPTIONS ('force' = 'false');
```
