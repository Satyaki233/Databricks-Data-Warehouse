```sql
-- Create table
CREATE TABLE IF NOT EXISTS bronze.customer (
  c_customer_sk INTEGER,
  c_customer_id STRING,
  c_current_cdemo_sk INTEGER,
  c_current_hdemo_sk INTEGER,
  c_current_addr_sk INTEGER,
  c_first_shipto_date_sk INTEGER,
  c_first_sales_date_sk INTEGER,
  c_salutation STRING,
  c_first_name STRING,
  c_last_name STRING,
  c_preferred_cust_flag STRING,
  c_birth_day INTEGER,
  c_birth_month INTEGER,
  c_birth_year INTEGER,
  c_birth_country STRING,
  c_login STRING,
  c_email_address STRING,
  c_last_review_date_sk INTEGER
)
USING DELTA;

-- Incremental load
COPY INTO bronze.customer
FROM (
  SELECT
    CAST(_c0 AS INTEGER) AS c_customer_sk,
    CAST(_c1 AS STRING) AS c_customer_id,
    CAST(_c2 AS INTEGER) AS c_current_cdemo_sk,
    CAST(_c3 AS INTEGER) AS c_current_hdemo_sk,
    CAST(_c4 AS INTEGER) AS c_current_addr_sk,
    CAST(_c5 AS INTEGER) AS c_first_shipto_date_sk,
    CAST(_c6 AS INTEGER) AS c_first_sales_date_sk,
    CAST(_c7 AS STRING) AS c_salutation,
    CAST(_c8 AS STRING) AS c_first_name,
    CAST(_c9 AS STRING) AS c_last_name,
    CAST(_c10 AS STRING) AS c_preferred_cust_flag,
    CAST(_c11 AS INTEGER) AS c_birth_day,
    CAST(_c12 AS INTEGER) AS c_birth_month,
    CAST(_c13 AS INTEGER) AS c_birth_year,
    CAST(_c14 AS STRING) AS c_birth_country,
    CAST(_c15 AS STRING) AS c_login,
    CAST(_c16 AS STRING) AS c_email_address,
    CAST(_c17 AS INTEGER) AS c_last_review_date_sk
  FROM 's3://satyaki-databricks-lakehouse/tpcds/landing/data/customer.dat'
)
FILEFORMAT = CSV
FORMAT_OPTIONS (
  'delimiter' = '|',
  'header' = 'false'
)
COPY_OPTIONS (
  'force' = 'false'
);
```

```sql
CREATE TABLE IF NOT EXISTS bronze.call_center (
    cc_call_center_sk         INTEGER,
    cc_call_center_id         STRING,
    cc_rec_start_date         DATE,
    cc_rec_end_date           DATE,
    cc_closed_date_sk         INTEGER,
    cc_open_date_sk           INTEGER,
    cc_name                   STRING,
    cc_class                  STRING,
    cc_employees              INTEGER,
    cc_sq_ft                  INTEGER,
    cc_hours                  STRING,
    cc_manager                STRING,
    cc_mkt_id                 INTEGER,
    cc_mkt_class              STRING,
    cc_mkt_desc               STRING,
    cc_market_manager         STRING,
    cc_division               INTEGER,
    cc_division_name          STRING,
    cc_company                INTEGER,
    cc_company_name           STRING,
    cc_street_number          STRING,
    cc_street_name            STRING,
    cc_street_type            STRING,
    cc_suite_number           STRING,
    cc_city                   STRING,
    cc_county                 STRING,
    cc_state                  STRING,
    cc_zip                    STRING,
    cc_country                STRING,
    cc_gmt_offset             DECIMAL(5,2),
    cc_tax_percentage         DECIMAL(5,2)
)
USING DELTA;

COPY INTO bronze.call_center
FROM (
    SELECT
        CAST(_c0  AS INTEGER)      AS cc_call_center_sk,
        CAST(_c1  AS STRING)       AS cc_call_center_id,
        CAST(_c2  AS DATE)         AS cc_rec_start_date,
        CAST(_c3  AS DATE)         AS cc_rec_end_date,
        CAST(_c4  AS INTEGER)      AS cc_closed_date_sk,
        CAST(_c5  AS INTEGER)      AS cc_open_date_sk,
        CAST(_c6  AS STRING)       AS cc_name,
        CAST(_c7  AS STRING)       AS cc_class,
        CAST(_c8  AS INTEGER)      AS cc_employees,
        CAST(_c9  AS INTEGER)      AS cc_sq_ft,
        CAST(_c10 AS STRING)       AS cc_hours,
        CAST(_c11 AS STRING)       AS cc_manager,
        CAST(_c12 AS INTEGER)      AS cc_mkt_id,
        CAST(_c13 AS STRING)       AS cc_mkt_class,
        CAST(_c14 AS STRING)       AS cc_mkt_desc,
        CAST(_c15 AS STRING)       AS cc_market_manager,
        CAST(_c16 AS INTEGER)      AS cc_division,
        CAST(_c17 AS STRING)       AS cc_division_name,
        CAST(_c18 AS INTEGER)      AS cc_company,
        CAST(_c19 AS STRING)       AS cc_company_name,
        CAST(_c20 AS STRING)       AS cc_street_number,
        CAST(_c21 AS STRING)       AS cc_street_name,
        CAST(_c22 AS STRING)       AS cc_street_type,
        CAST(_c23 AS STRING)       AS cc_suite_number,
        CAST(_c24 AS STRING)       AS cc_city,
        CAST(_c25 AS STRING)       AS cc_county,
        CAST(_c26 AS STRING)       AS cc_state,
        CAST(_c27 AS STRING)       AS cc_zip,
        CAST(_c28 AS STRING)       AS cc_country,
        CAST(_c29 AS DECIMAL(5,2)) AS cc_gmt_offset,
        CAST(_c30 AS DECIMAL(5,2)) AS cc_tax_percentage
    FROM 's3://satyaki-databricks-lakehouse/tpcds/landing/data/call_center.dat'
)
FILEFORMAT = CSV
FORMAT_OPTIONS ('delimiter' = '|', 'header' = 'false')
COPY_OPTIONS ('force' = 'false');
```
