```sql
%sql
CREATE TABLE IF NOT EXISTS silver.store_sales (
    ss_sold_date_sk          BIGINT,
    ss_sold_time_sk          BIGINT,
    ss_item_sk               BIGINT,
    ss_customer_sk           BIGINT,
    ss_cdemo_sk              BIGINT,
    ss_hdemo_sk              BIGINT,
    ss_addr_sk               BIGINT,
    ss_store_sk              BIGINT,
    ss_promo_sk              BIGINT,
    ss_ticket_number         BIGINT,
    ss_quantity              INT,
    ss_wholesale_cost        DECIMAL(7,2),
    ss_list_price            DECIMAL(7,2),
    ss_sales_price           DECIMAL(7,2),
    ss_ext_discount_amt      DECIMAL(7,2),
    ss_ext_sales_price       DECIMAL(7,2),
    ss_ext_wholesale_cost    DECIMAL(7,2),
    ss_ext_list_price        DECIMAL(7,2),
    ss_ext_tax               DECIMAL(7,2),
    ss_coupon_amt            DECIMAL(7,2),
    ss_net_paid              DECIMAL(7,2),
    ss_net_paid_inc_tax      DECIMAL(7,2),
    ss_net_profit            DECIMAL(7,2)
)
USING DELTA
PARTITIONED BY (ss_sold_date_sk);

MERGE INTO silver.store_sales AS target
USING (
    SELECT * FROM bronze.store_sales
    WHERE ss_item_sk IS NOT NULL AND ss_ticket_number IS NOT NULL
) AS source
ON  target.ss_item_sk       = source.ss_item_sk
AND target.ss_ticket_number = source.ss_ticket_number
WHEN NOT MATCHED THEN INSERT *;
```

```sql
%sql
CREATE TABLE IF NOT EXISTS silver.inventory (
    inv_date_sk          BIGINT,
    inv_item_sk          BIGINT,
    inv_warehouse_sk     BIGINT,
    inv_quantity_on_hand INT
)
USING DELTA
CLUSTER BY (inv_date_sk);

MERGE INTO silver.inventory AS target
USING (
    SELECT * FROM bronze.inventory
    WHERE inv_date_sk IS NOT NULL AND inv_item_sk IS NOT NULL
) AS source
ON  target.inv_date_sk      = source.inv_date_sk
AND target.inv_item_sk      = source.inv_item_sk
AND target.inv_warehouse_sk = source.inv_warehouse_sk
WHEN MATCHED AND
    target.inv_quantity_on_hand IS DISTINCT FROM source.inv_quantity_on_hand
THEN UPDATE SET *
WHEN NOT MATCHED THEN INSERT *;
```
